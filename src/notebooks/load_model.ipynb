```python
from tensorflow.keras.models import model_from_json

def fix_input_layer(json_path, weights_path):
    # Load the model architecture from JSON
    with open(json_path, "r") as file:
        model_json = file.read()

    # Replace "batch_shape" with "input_shape" in the JSON structure
    fixed_json = model_json.replace('"batch_shape":', '"input_shape":')

    # Create the model from the fixed JSON
    model = model_from_json(fixed_json)

    # Load the weights into the model
    model.load_weights(weights_path)

    return model

# Paths to architecture JSON and weights file
json_path = "model_architecture.json"
weights_path = "../../models/hybrid_model.weights.h5"

# Fix and load the model
ensemble_model = fix_input_layer(json_path, weights_path)

# Verify the model
print("Model loaded successfully.")
print(ensemble_model.summary())


# # Fix and load the model
# json_path = "model_architecture.json"
# weights_path = "../../models/hybrid_model.weights.h5"
# ensemble_model = fix_input_layer(json_path, weights_path)

# print("Model loaded successfully.")
```
